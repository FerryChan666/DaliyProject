# 3.数据可视化

## 统计每个国家的大学数量
使用df的函数groupby('region')，将有相同region的记录合并成一个表格（E.g.中国是一个表格，美国是一个表格...)。再用count()记录每个表格有多少条记录（有多少个大学），最后在dataframe中用[]（意思是，只获取一列数据）

```python
df_counts=df.groupby('country').count()['name']
```

```python
df.groupby('country').count()['name']
```

## 求每个国家的大学平均分

```python
df_means=df.groupby('country')['score'].mean()
```

```python
df.groupby('country')['score'].mean()
```

把index作为x轴，values作为y轴

```python
from pyecharts import options as opts
```

```python
name=df_counts.index.tolist()
counts=df_counts.values.tolist()
means=df_means.values.astype("int64").tolist()

bar=(
    Bar()
    .add_xaxis(name)
    .add_yaxis('counts', counts)
    .add_yaxis('means', means)
)
 
bar.render_notebook()
```

```python
m=(
    Map()

    .add(series_name="Map", data_pair=[(i,j) for i,j in zip(name,counts)], maptype="world", 
         is_map_symbol_show=False, name_map=name_map.name_translation)
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(max_=42)
    
    )
)
m.render_notebook()
```
